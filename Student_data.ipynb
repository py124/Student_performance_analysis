```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('Student_performance_data _.csv')
print(df.head())
print(df.info())
print(df.describe())
```

```python
plt.bar(df['GradeClass'], df['GPA'], color = 'green')
plt.xlabel('Grade Class')
plt.ylabel('GPA')
plt.title('Grade Class-wise GPA')
plt.grid(True)
plt.show()
```

# **Regression**

```python
x = df[['StudyTimeWeekly']]
y = df['GPA']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"RÂ² Score: {r2_score(y_test, y_pred)}")
```

```python
plt.scatter(x_test, y_test, color='green', label='Actual')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel("Weekly Study Time")
plt.ylabel("GPA")
plt.title(" Relationship between Study time and GPA")
plt.legend()
plt.grid(True)
plt.show()
```

####**The linear regression model attempts to explain the relationship between weekly study time and GPA. Here's an interpretation of the results:**

**Intercept:** The intercept of 1.596 suggests that, according to this model, a student who studies 0 hours per week is predicted to have a GPA of approximately 1.60.

**Coefficient:** The coefficient of 0.031 indicates that for every additional hour of weekly study time, the model predicts an increase in GPA of approximately 0.031. This suggests a weak positive relationship between study time and GPA.

**Mean Squared Error (MSE):** The MSE of 0.821 measures the average of the squared differences between the actual GPA values and the predicted GPA values. A lower MSE indicates a better fit of the model to the data. In this case, the MSE is relatively high, suggesting that the model's predictions are not very close to the actual values.

**R-squared Score:** The R-squared score of 0.0075 indicates that only about 0.75% of the variance in GPA can be explained by weekly study time according to this model. This is a very low R-squared value, suggesting that weekly study time alone is not a strong predictor of GPA in this dataset.

The scatter plot visually confirms these findings. The points are widely scattered, and the regression line (red) is nearly horizontal, indicating a very weak positive correlation between weekly study time and GPA. This suggests that other factors not included in this simple linear regression model likely play a much larger role in determining a student's GPA.
